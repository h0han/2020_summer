# cancer (실전에서 많이 응용되는 코드이므로 숙달하도록 하자)

```python
cancer['data'].shape # cancer.data
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, stratify = cancer.target, random_state = 66
)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
train_accuracy = []
test_accuracy = []
for n_neighbors in range(1, 11):
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    train_accuracy.append(clf.score(X_train, y_train))  # 훈련 정확도를 append
    test_accuracy.append(clf.score(X_test, y_test))  # 테스트 정확도를 append
print(train_accuracy)
print(test_accuracy)
```

```python
plt.plot(range(1, 11), train_accuracy, label = 'train accuracy')
plt.plot(range(1, 11), test_accuracy, label = 'test accuracy')
plt.legend()
plt.show()
```

# LogisticRegression 규제
# SVM 규제 : C; cost

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify = cancer.target, random_state = 7)
logreg = LogisticRegression().fit(X_train, y_train)
print("train accuracy: {:.3f}".format(logreg.score(X_train, y_train)))
print("test accuracy: {:.3f}".format(logreg.score(X_test, y_test)))
```

```python
logreg = LogisticRegression(C = 100).fit(X_train, y_train)
print("train accuracy: {:.3f}".format(logreg.score(X_train, y_train)))
print("test accuracy: {:.3f}".format(logreg.score(X_test, y_test)))
```

```python
logreg = LogisticRegression(C = 0.01).fit(X_train, y_train)
print("train accuracy: {:.3f}".format(logreg.score(X_train, y_train)))
print("test accuracy: {:.3f}".format(logreg.score(X_test, y_test)))
```

```python
logreg = LogisticRegression(C = 0.0001).fit(X_train, y_train)
print("train accuracy: {:.3f}".format(logreg.score(X_train, y_train)))
print("test accuracy: {:.3f}".format(logreg.score(X_test, y_test)))
```

```python
for C, marker in zip([0.001, 1, 100], ['o', '^', 'v']):
    lr_l1 = LogisticRegression(solver = 'liblinear', C=C, penalty = 'l1').fit(X_train, y_train)
    print("C:{:.3f}, 훈련 정확도:{:.2f}".format(C, lr_l1.score(X_train, y_train)))
    print("C:{:.3f}, 테스트 정확도:{:.2f}".format(C, lr_l1.score(X_test, y_test)))
```

```python
linear_svm = LinearSVC().fit(X, y)
```

```python
print("기울기 배열의 크기:", linear_svm.coef_.shape)
print("절편 배열의 크기:", linear_svm.intercept_.shape)
```

```python
linear_svm.intercept_
```

```python
import platform
import matplotlib
if platform.system() == 'Windows':
    matplotlib.rc('font', family = 'Malgun Gothic')
elif platform.system() == 'Darwin':  #Mac OS
    matplotlib.rc('font', family = 'AppleGothic')
elif platform.system() == 'Linux':  #Linux
    matplotlib.rc('font', family = 'NanumGothic')
    
plt.rcParams['axes.unicode_minus'] = False  # minus 기호 깨짐 방지
```

```python
mglearn.plots.plot_2d_classification(linear_svm, X, fill=True, alpha=.7)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
line = np.linspace(-15, 15)
for coef, intercept, color in zip(linear_svm.coef_, linear_svm.intercept_,
                                  mglearn.cm3.colors):
    plt.plot(line, -(line * coef[0] + intercept) / coef[1], c=color)
plt.legend(['클래스 0', '클래스 1', '클래스 2', '클래스 0 경계', '클래스 1 경계',
            '클래스 2 경계'], loc=(1.01, 0.3))
plt.xlabel("특성 0")
plt.ylabel("특성 1")
```

```python
from sklearn.tree import DecisionTreeClassifier
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify = cancer.target, random_state = 42)
tree = DecisionTreeClassifier(random_state = 0)
tree.fit(X_train, y_train)
print("훈련 정확도:{:.3f}".format(tree.score(X_train, y_train)))
print("테스트 정확도:{:.3f}".format(tree.score(X_test, y_test)))
# 훈련 정확도:1.000 ==> OVERFITTING 의사결정 tree는 내부적으로 depth 탐색을 하는데, 이 탐색 경로를 인지하고 있음.
# 테스트 정확도:0.937 
```

```python
tree = DecisionTreeClassifier(max_depth = 4, random_state = 0)  # 4칸까지만 학습 후 암기할 수 있고 이후는 암기하지 않음 ==> overfitting 방지
tree.fit(X_train, y_train)
print("훈련 정확도:{:.3f}".format(tree.score(X_train, y_train)))
print("테스트 정확도:{:.3f}".format(tree.score(X_test, y_test)))
```

```python
from sklearn.tree import export_graphviz
```

```python
export_graphviz(tree, out_file="tree.dot", class_names=['악성', '양성'], feature_names=cancer.feature_names, impurity = False, filled = True)  # tree.dot이라는 파일 생성
import graphviz
with open('tree.dot', encoding = 'UTF8') as f:
    dot_graph = f.read()
display(graphviz.Source(dot_graph))
```

```python
print("특성 중요도\n", tree.feature_importances_)
```

```python
plt.figure(figsize = (15, 10))
def plot_feature_importances_cancer(model):
    n_features = cancer.data.shape[1]
    plt.barh(np.arange(n_features), model.feature_importances_, align='center')  # barchar, horizontal
    plt.yticks(np.arange(n_features), cancer.feature_names)
    plt.xlabel("특성 중요도")
plot_feature_importances_cancer(tree)
plt.show()
```
