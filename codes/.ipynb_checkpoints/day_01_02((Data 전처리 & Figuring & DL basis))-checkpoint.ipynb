```python
#ML의 경우 dataset을 0, 1로 바꾸어야 하기 때문에 이를 위한 처리
tmp = []
for i in df['sex']:
    if i == 'female': tmp.append(1)
    else: tmp.append(0)
df['sex'] = tmp
df
```

```python
#결측치에 대한 처리 - 결측치 버림
df = df[df['age'].notnull()]
df.info()
```

```python
df[df['fare'].isnull()]
```

```python
df = df[df['fare'].notnull()]
df.info()
```

```python
#feature selection
#         pclass, sex, age, slisp, parch, fare
# index : 0       3    4    5      6      8
x_data = df.values[:, [0,3,4,5,6,8]]
x_data
```

```python
y_data = df.values[:, [1]]  #1 is 'survived'
y_data
```

```python
import random
random.seed(7)
random.randint(1, 100)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    x_data, y_data, test_size = 0.1, random_state = 7)  # train data : 90%, test data : 10%
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
X_train[0]
```

```python
y_train[0]
```

```python
import keras
from sklearn.model_selection import train_test_split
```

```python
from keras.models import Sequential  # struct
from keras.layers import Dense  # each layer
```

```python
model = Sequential()
```

```python
model.add(Dense(128, input_shape = (6,), activation = 'relu'))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
model.compile(loss = 'mse', optimizer = 'adam', metrics = ['accuracy'])  # Generally, when using sigmoid func, using mse loss func
```

```python
#%%time  
hist = model.fit(X_train, y_train, validation_data = (X_test, y_test), epochs = 500)  ## epochs; training 
```

```python
plt.figure(figsize = (12, 8))
plt.plot(hist.history['loss'])  # 훈련 비용
plt.plot(hist.history['val_loss'])  # 검증 비용
plt.plot(hist.history['acc'])  # 훈련 정확도
plt.plot(hist.history['val_acc'])  # 검증 정확도
plt.legend(['loss', 'val_loss', 'acc', 'val_acc'])
plt.show()
```

```python
# 디카프리오는 살 수 있었나?
# 3등실, 남성:0, 19살, 형제 x:0, 부모 x:0, fare:5달러
dicaprio = np.array([3, 0, 19, 0, 0, 5]).reshape(1, 6) # 2차원으로 만듦
model.predict(dicaprio)
# array([[0.01050257]], dtype=float32), 생존확률 1%
```

```python
# 윈슬릿의 경우
# 1등실, 여성:1, 17살, 형제:1, 부모:2, fare:100달러
winslet = np.array([1, 1, 17, 1, 2, 100]).reshape(1, 6)
model.predict(winslet)
# array([[0.9999995]], dtype=float32), 생존확률 99.9%
```
